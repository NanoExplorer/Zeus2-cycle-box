```python
curr=np.array(s.sensors['Current'][0])
plot(curr[:,0],curr[:,1],'.')
```

```python
from scipy.stats import linregress
numpts=14
x=curr[-numpts:,0]
y=curr[-numpts:,1]
plot(x,y,'.')
res=linregress(x,y)
plot(x,x*res.slope+res.intercept)
```

```python
print(9/res.slope/60/60)
print(res.slope)
```
