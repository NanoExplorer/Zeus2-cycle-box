```python
from scipy.stats import linregress
```

```python
t = database.EasyThermometry(2500,want_sensors="magnet")
```

```python
current=np.array(t.sensors['Current'][0])
```

```python
start=-120
end=-1
res=linregress(current[start:end,0],current[start:end,1])
print(res.slope)
print(5/res.slope/60/60)
plt.plot(current[:,0],current[:,1])
plt.plot(current[start:end,0],current[start:end,0]*res.slope+res.intercept)
grid(True)
grid(True,which='minor',linestyle='-', linewidth=0.08,color='k')
minorticks_on()
```
