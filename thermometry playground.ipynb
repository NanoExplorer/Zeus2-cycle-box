```python
from scipy.stats import linregress
```

```python
t = database.EasyThermometry(5000,want_sensors="magnet")
```

```python
current=np.array(t.sensors['Current'][0])
```

```python
res=linregress(current[:,0],current[:,1])
print(9/res.slope/60/60)
print(res.slope*60*60)
plt.plot(current[:,0],current[:,1])
plt.plot(current[:,0],current[:,0]*res.slope+res.intercept)
grid(True)
grid(True,which='minor',linestyle='-', linewidth=0.08,color='k')
minorticks_on()
```
